```python
import pandas as pd
import io
from google.colab import files

class DataPrepKit:
    def __init__(self, data):
        self.data = data

    def read_data(self, file_path):
        if file_path.endswith('.csv'):
            return pd.read_csv(file_path)
        elif file_path.endswith('.xlsx'):
            return pd.read_excel(file_path)
        elif file_path.endswith('.json'):
            return pd.read_json(file_path)
        else:
            print("Unsupported file type")
            return None

    def generate_summary(self):
        return self.data.describe(include='all')

    def handle_missing_values(self, strategy):
        if strategy == 'drop':
            return self.data.dropna()
        elif strategy == 'mean':
            return self.data.fillna(self.data.mean())
        else:
            print("Unsupported strategy")
            return None

    def encode_categorical_data(self):
        return pd.get_dummies(self.data, drop_first=True)

# Prompt user to upload file
print("Please upload your data file:")
uploaded = files.upload()

# Determine file type and read data
file_path = next(iter(uploaded))
data_prep = DataPrepKit(pd.read_csv(io.BytesIO(uploaded[file_path]))) if file_path.endswith('.csv') else DataPrepKit(pd.read_excel(io.BytesIO(uploaded[file_path]))) if file_path.endswith('.xlsx') else DataPrepKit(pd.read_json(io.BytesIO(uploaded[file_path])))

# Generate summary
summary = data_prep.generate_summary()

# Handle missing values
df_cleaned = data_prep.handle_missing_values(strategy='mean')

# Encode categorical data
df_encoded = data_prep.encode_categorical_data()
```
